```python
"""
script_evaluators_retrieval.py
--------------------------------
Script-based RETRIEVAL evaluators for the dataset-ecli experiment in Langfuse.
Runs across all three model conditions: GPT-4o, Claude 3.5 Sonnet, Gemini 1.5 Pro.

IMPORTANT - WHAT THIS MEASURES:
    This script evaluates RETRIEVAL quality, not citation accuracy.
    It compares the ECLIs the RETRIEVER surfaced (trace.output.retrievedDocs)
    against the expected ECLIs in each dataset item's expectedOutput.Sources.

    This is NOT the same as asking "did the model cite the right sources in
    its answer". A document can be retrieved (appear in retrievedDocs) without
    ever being used or cited in the final generated answer. This script can
    only tell you whether the right case law was *available* to the model,
    not whether the model actually *used* it correctly.

    Citation-level accuracy (script_evaluators_ecli.py) remains BLOCKED:
    the pipeline does not currently log the model's final answer or the
    ECLIs it cited anywhere in Langfuse. Do not relabel these retrieval
    scores as "citation accuracy" or "answer accuracy" -- they measure a
    different, earlier stage of the pipeline (retrieval), not generation.

Metrics (all 0-1), computed over retrievedDocs vs expected Sources:
  - retrieval_precision : fraction of RETRIEVED ECLIs that were correct
  - retrieval_recall     : fraction of EXPECTED ECLIs that were retrieved
  - retrieval_hit        : 1.0 if at least one expected ECLI was retrieved, else 0.0

Requirements:
    pip install langfuse
"""

from langfuse import Langfuse

DATASET_NAME = "dataset-eclis-final"

RUN_NAMES = [
    "saga-ecli-gpt",
    "saga-ecli-claude",
    "saga-ecli-gemini",
]



def extract_expected_eclis(expected_output) -> set[str]:
    if expected_output is None:
        return set()
    if isinstance(expected_output, dict):
        inner = expected_output.get("expectedOutput", expected_output)
        sources = inner.get("Sources", inner.get("sources", []))
        if isinstance(sources, list):
            return {s.strip() for s in sources if isinstance(s, str)}
    return set()


def extract_retrieved_eclis(trace_output) -> set[str]:
    if not isinstance(trace_output, dict):
        return set()
    docs = trace_output.get("retrievedDocs", [])
    if not isinstance(docs, list):
        return set()
    return {
        d.get("documentId", "").strip()
        for d in docs
        if isinstance(d, dict) and d.get("documentId")
    }




def retrieval_precision(retrieved: set[str], expected: set[str]) -> float:
    if not retrieved:
        return 0.0
    return len(retrieved & expected) / len(retrieved)


def retrieval_recall(retrieved: set[str], expected: set[str]) -> float:
    if not expected:
        return 0.0
    return len(retrieved & expected) / len(expected)


def retrieval_hit(retrieved: set[str], expected: set[str]) -> float:
    if not expected:
        return 0.0
    return 1.0 if (retrieved & expected) else 0.0



def push_score(client: Langfuse, trace_id: str, name: str, value: float, run_name: str):
    client.create_score(
        trace_id=trace_id,
        name=f"{name}",  
        value=round(value, 4),
        data_type="NUMERIC",
    )




def evaluate_run(client: Langfuse, dataset, run_name: str):
    print(f"\n{'═' * 60}")
    print(f"  Run: {run_name}")
    print(f"{'═' * 60}\n")

    try:
        run = client.api.datasets.get_run(DATASET_NAME, run_name)
    except Exception as e:
        print(f"[ERROR] Could not fetch run '{run_name}': {e}")
        return []

    item_to_trace: dict[str, str] = {}
    for run_item in run.dataset_run_items:
        if run_item.trace_id:
            item_to_trace[run_item.dataset_item_id] = run_item.trace_id

    print(f"Dataset : {DATASET_NAME}  ({len(dataset.items)} items)")
    print(f"Run     : {run_name}  ({len(item_to_trace)} linked traces)\n")

    results = []

    for item in dataset.items:
        trace_id = item_to_trace.get(item.id)
        if not trace_id:
            print(f"[SKIP] Item {item.id} — no linked trace in run '{run_name}'")
            continue

        expected_eclis = extract_expected_eclis(item.expected_output)
        query = item.input.get("input", {}).get("query", "")

        if not expected_eclis:
            print(f"[SKIP] trace={trace_id[:8]}…  query={query[:50]!r} — no expected ECLIs, excluded from average")
            continue

        try:
            trace_data = client.api.trace.get(trace_id)
            retrieved_eclis = extract_retrieved_eclis(trace_data.output)
        except Exception as e:
            print(f"[ERROR] Could not fetch trace {trace_id}: {e}")
            continue

        print(f"Scoring  trace={trace_id[:8]}…  query={query[:50]!r}")
        print(f"  expected ECLIs  : {sorted(expected_eclis)}")
        print(f"  retrieved ECLIs : {len(retrieved_eclis)} docs")

        p   = retrieval_precision(retrieved_eclis, expected_eclis)
        r   = retrieval_recall(retrieved_eclis, expected_eclis)
        hit = retrieval_hit(retrieved_eclis, expected_eclis)

        push_score(client, trace_id, "retrieval_precision", p, run_name)
        push_score(client, trace_id, "retrieval_recall",     r, run_name)
        push_score(client, trace_id, "retrieval_hit",        hit, run_name)

        results.append({
            "trace_id": trace_id,
            "retrieval_precision": p,
            "retrieval_recall": r,
            "retrieval_hit": hit,
        })

        print(f"  retrieval_precision={p:.3f}  retrieval_recall={r:.3f}  retrieval_hit={hit:.3f}\n")

    return results



def run_evaluators():
    client = Langfuse()
    client.auth_check()
    print("Connected to Langfuse ✓\n")

    dataset = client.get_dataset(DATASET_NAME)
    all_results: dict[str, list] = {}

    for run_name in RUN_NAMES:
        results = evaluate_run(client, dataset, run_name)
        all_results[run_name] = results

    print(f"\n{'═' * 60}")
    print("  Cross-run Summary")
    print(f"{'═' * 60}\n")
    print(f"  {'Run':<25} {'Precision':>10} {'Recall':>10} {'Hit Rate':>10} {'N':>5}")
    print(f"  {'-'*25} {'-'*10} {'-'*10} {'-'*10} {'-'*5}")

    for run_name, results in all_results.items():
        if not results:
            print(f"  {run_name:<25} {'—':>10} {'—':>10} {'—':>10} {'0':>5}")
            continue
        n = len(results)
        avg_p   = sum(r["retrieval_precision"] for r in results) / n
        avg_r   = sum(r["retrieval_recall"]    for r in results) / n
        avg_hit = sum(r["retrieval_hit"]       for r in results) / n
        print(f"  {run_name:<25} {avg_p:>10.4f} {avg_r:>10.4f} {avg_hit:>10.4f} {n:>5}")

    print(
        "\nReminder: these are RETRIEVAL metrics (was the right case law\n"
        "surfaced by the retriever), not citation/answer accuracy metrics.\n"
        "Citation accuracy remains blocked pending pipeline instrumentation."
    )

    client.flush()
    print("\nDone — scores pushed to Langfuse ✓")


if __name__ == "__main__":
    run_evaluators()
```
